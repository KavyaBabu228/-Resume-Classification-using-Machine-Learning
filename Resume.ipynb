```python
import os
import textract
import pandas as pd
```

```python
folder_path = "/content/drive/MyDrive/Resumes_Docx"
print(os.listdir(folder_path))
```

```python
file_path1 = []
category1 = []

directory1 = '/content/drive/MyDrive/Resumes_Docx/workday'

# Check if the directory exists before proceeding
if os.path.exists(directory1):
    for i in os.listdir(directory1):
        if i.endswith('.docx'):
            file_path = os.path.join(directory1, i)
            try:
                text = textract.process(file_path).decode('utf-8')
                file_path1.append(text)
                category1.append('workday')
            except Exception as e:
                print(f"Error processing {file_path}: {e}")
else:
    print(f"Directory '{directory1}' not found. Please check the path.")

# Create a DataFrame
data1 = pd.DataFrame({'Raw_Details': file_path1, 'Category1': category1})

# Display the DataFrame
print(data1)
```
